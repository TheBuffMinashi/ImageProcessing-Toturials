# Import all necessary libraries in the following cell:

```python
import cv2
from google.colab.patches import cv2_imshow
```

# Create functions for all parts of the process:

```python
# Create a function to read an image and display its info and dimensions

def read_and_display(filename):

  # Read the image
  image = cv2.imread(filename)

  # Get dimensions of image
  dimensions = image.shape
  
  # Height, width, number of channels in image
  height = image.shape[0]
  width = image.shape[1]
  channels = image.shape[2]
  
  print('Image Dimension    : ',dimensions)
  print('Image Height       : ',height)
  print('Image Width        : ',width)
  print('Number of Channels : ',channels)
  print('\n')
  cv2_imshow(image)
```

```python
filename = str(fpath) + str('Flower.jpg')
# print(filename)
read_and_display(filename)
```
